## Categories of Data Science Tools

Open source tools are available for various data science tasks.

**Data Management** is the process of persisting and retrieving data.


**Data Integration and Transformation** (ETL) is the process of retrieving data from remote data management systems.

**Data Visualization** is part of both the initial data exploration process, as well as being part of the final deliverable.

**Model Building** is the process of creating a machine learning or deep learnign model. This model uses the appropriate algorithm trained with a lot of data

**Model Deployment** makes the machine learning model available to 3rd-party applications.

**Model Monitoring & Assessment** ensures continuous performance quality checks on the deployed models.

**Code Asset Management** uses versioning and other collaborative features to facilitate teamwork

**Data Asset Management** brings the same versioning and collaborative components to data. It also supports replication, backup, and access right management.

**IDEs** are tools that help data scientists to implement, execute, tests, and deploy their work.

**Execution Environments** are tools where data preprocessing, model training, and deployment take place.

## Open Source Tools for Data Science

The most widely-used data management tools are **Relational Databases**. This includes:
* MySQL
* PostgreSQL

NoSQL databases are also great for data management e.g.
* MongoDB
* Apache CouchDB
* Apache Cassandra

Along with File-based tools and cloud file systems:
* Hadoop
* Ceph
* S3

ETL is often proposed to be changed to ELT. This stresses the fact that data is dumped somewhere and that the Data Engineer or Data Scientist is responsible for this data.

Tools for ETL include:
* Apache Airflow
* Apache Kafka
* Spark SQL
* Kubeflow
* Apache nifi
* Node-RED (consumes very little resources, can run on a rasberry pi)

**Data Visualization** tools:
* Many libraries
* Hue
* kibana (EDA web app limited to elastisearch)

**Model Deployment** - once you've created a machine learning model capable of predicting some key aspects of the future, one should make that model consumable by other developers and turn it into an API. Some of the ways to do this include:
* Apache PredictionIO (only supports Apache Spark ML models)
* Seldon (supports nearly every framework like TensorFlow, SparkML, R, Scikit-Learn)
* MLeap (SparkML)
* Tensorflow Service (Only tensorflow models, includes tensorflow.js)

**Model Monitoring and Assessment** tools:
* ModelDB
* Prometheus
* IBM AI Fairnes 360 Open Source Toolkit (good for detecting bias in ML models)
* IBM Adversarial Robustness 360 Toolbox (good for making a model most robust)
* IBM Explainability 360 Toolkit - presents similar datapoints within a dataset for manual comparison
