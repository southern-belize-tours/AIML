## Two Main Types

**Content-Based** vs. **Collaborative-Filtering**

**Content-Based** systems try to figure out which aspects a user likes about something and shows them more of the same of what they like.

**Collaborative Filtering** is based on a user saying "*tell me what's popular among my neighbors because I might like it too*". It finds similar groups of users and provides recommendations based on similar tastes within that group.

Two types of recommender systems:
* Memory Based: uses the entire user-item dataset to generate recommendation. Looks aat things like Correlation, Cosine Similarity, and Euclidean Distance.
* Model-based: A model of users is developed in an attempt to learn their preferences. ML techniques like regression, clustering, and classification are leveraged.
